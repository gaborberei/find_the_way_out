# SELENIUM SOLVER

```python
def solver(adat):
    
    """
    ####################################################################
                           Cleaning the data
    ####################################################################
    """
    
    for sor in range(len(adat)):
        for db in range(len(adat[0])):
            if adat[sor][db] == " ":
                adat[sor][db] = 1

    for sor in range(len(adat)):
        for db in range(len(adat[0])):
            adat[sor][db] = int(adat[sor][db])    
    
    """
    ####################################################################
                        Get the position of the doors
    ####################################################################
    """
    for sor in range(len(adat)):
        for db in range(len(adat[0])):
            if adat[sor][db] == 2:
                x = sor
                y = db
                adat[sor][db] = 0
                #print("y =", sor, "x =", db)
            if adat[sor][db] == 3:
                a = sor
                b = db

                adat[sor][db] = "OUT"
                #print("a =",sor,"b =",db)
    """
    ####################################################################
                                    MOVING
    ####################################################################
    """
    moving = []
    elozo_pont = []

    while not ("END" in moving):

        R = adat[x+1][y]
        L = adat[x-1][y] 
        U = adat[x][y+1]
        D = adat[x][y-1]

        try:
            if R+L+U+D>1:
                k = [x,y]
                elozo_pont.append(k)
                moving.append("Y")

            if R==0 and L==0 and U==0 and D==0:
                x=elozo_pont[-1][0]
                y=elozo_pont[-1][1]
                elozo_pont.remove(elozo_pont[-1])
                index = list(reversed(moving)).index("Y")

                del moving[len(moving)-index-1:]

        except:
            print("")


        if L == 1:
            adat[x][y] = 0
            moving.append("U") #L
            x = x-1

        elif R == 1:
            adat[x][y] = 0
            x = x+1
            moving.append("D") #R

        elif U == 1:
            adat[x][y] = 0
            moving.append("R") #U
            y = y+1

        elif D == 1:
            adat[x][y] = 0
            moving.append("L") #D
            y = y-1

    ####################################################################     

        if R == "OUT":
            adat[x][y] = 0
            moving.append("D")
            print("END --- KEY IS:")
            moving.append("END")

        elif U == "OUT":
            adat[x][y] = 0
            moving.append("R")
            print("END --- KEY IS:")
            moving.append("END")

        elif L == "OUT":
            adat[x][y] = 0
            moving.append("U")
            print("END --- KEY IS:")
            moving.append("END")

        elif D == "OUT":
            adat[x][y] = 0
            moving.append("L")
            print("END --- KEY IS:")
            moving.append("END")

    ####################################################################

    moving = [f for f in moving if not f=="Y"]

    moving.remove("END")

    moving = "".join(moving)

    return moving
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, (len(lst)), n):
        yield lst[i:i + n]
```
